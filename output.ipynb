```python
import tensorflow as tf
import numpy as np
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
from tkinter import *
from tkinter import ttk

root = Tk()

model = load_model('model_regularized_better.keras')

with open('columns.pkl', 'rb') as file:
    columns = pickle.load(file)
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

def get_user_input(): 
    runs = float(run_in.get()) 
    wickets = float(wicket_in.get())
    balls = float(ball_in.get())
    target = float(target_in.get())
    last_5 = float(last5_in.get())
    batting =  batteam_select.get()
    bowling = ballteam_select.get()
    venue = venue_select.get()
    CRR = (runs/balls)*6
    RRR = ((target - runs)/(120 - balls))*6
    wickets_left = 10 - wickets
    balls_left = 120 - balls
    runs_left = target - runs
    user_data = {
        'Innings Runs': [runs],
        'Target Score': [target],
        'Last Five': [last_5],
        'CRR': [CRR],
        'RRR': [RRR],
        'Wickets Left': [wickets_left],
        'Balls_left': [balls_left],
        'Balls_bowled': [balls],
        'Runs_left': [runs_left],
        'Venue': [venue],
        'Bat Second': [batting],
        'Bat First': [bowling],
        
    } 
    return pd.DataFrame(user_data)

def predict_comm():
       user_input_df = get_user_input()
       user_input_encoded = pd.get_dummies(user_input_df)      
       user_input_encoded = user_input_encoded.reindex(columns=columns, fill_value=0)
       user_input_encoded = user_input_encoded*1
       user_input_scaled = scaler.transform(user_input_encoded)
       predictions = model.predict(user_input_scaled)
       prediction_lb.config(text=f'{predictions[0][0]*100}')

def update_opt(event):
    current_txt = venue_cb.get()
    current_opt = [i for i in venue_option if current_txt.lower() in i.lower()]
    venue_cb['values'] = current_opt

batteam_select = StringVar()
ballteam_select = StringVar()
venue_select = StringVar()


team_option = ['West Indies', 'England', 'Bangladesh', 'India', 'Pakistan',
       'Sri Lanka', 'Afghanistan', 'South Africa', 'Australia',
       'New Zealand']

venue_option = ['SuperSport Park', 'Bellerive Oval',
       'Pallekele International Cricket Stadium', 'R Premadasa Stadium',
       'Hagley Oval', 'The Rose Bowl', 'Coolidge Cricket Ground',
       'Dubai International Cricket Stadium',
       'Saurashtra Cricket Association Stadium',
       'Rajiv Gandhi International Cricket Stadium', 'Wankhede Stadium',
       'Beausejour Stadium', 'Arun Jaitley Stadium', 'Kensington Oval',
       'Manuka Oval', 'Sophia Gardens', 'Shere Bangla National Stadium',
       'Adelaide Oval', 'Trent Bridge', 'Newlands', 'Old Trafford',
       'Sharjah Cricket Stadium', 'Holkar Cricket Stadium',
       'Providence Stadium', 'Sydney Cricket Ground',
       'Sheikh Zayed Stadium', 'National Stadium',
       'New Wanderers Stadium', 'Eden Park', 'County Ground',
       'Riverside Ground', "Lord's", 'Eden Gardens', 'McLean Park',
       'National Cricket Stadium', 'Kennington Oval', 'Seddon Park',
       'Vidarbha Cricket Association Stadium', "St George's Park",
       'Zayed Cricket Stadium',
       'Mahinda Rajapaksa International Cricket Stadium', 'Warner Park',
       'The Wanderers Stadium', 'Maharashtra Cricket Association Stadium',
       'Melbourne Cricket Ground', 'Kingsmead', "Queen's Park Oval",
       'Punjab Cricket Association IS Bindra Stadium', 'Westpac Stadium',
       'Central Broward Regional Park Stadium Turf Ground',
       'M Chinnaswamy Stadium', 'Gaddafi Stadium', 'Stadium Australia',
       'Windsor Park', 'Perth Stadium', 'Bay Oval',
       'Narendra Modi Stadium',
       'Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium',
       'Brisbane Cricket Ground', 'Sabina Park', 'Barabati Stadium',
       'Edgbaston', 'Sylhet International Cricket Stadium',
       'Darren Sammy National Cricket Stadium', 'Harare Sports Club',
       'Himachal Pradesh Cricket Association Stadium',
       'R.Premadasa Stadium', 'Feroz Shah Kotla',
       'Zahur Ahmed Chowdhury Stadium',
       'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium',
       'Rawalpindi Cricket Stadium', 'JSCA International Stadium Complex',
       'Buffalo Park', 'Sir Vivian Richards Stadium',
       'Gymkhana Club Ground']

run_lb = Label(root,text='Enter Runs')
run_lb.grid(row=0, column=0)

wicket_lb = Label(root,text='Enter Wickets')
wicket_lb.grid(row=1, column=0)

ball_lb = Label(root,text='Enter Balls Bowled')
ball_lb.grid(row=2, column=0)

target_lb = Label(root,text='Enter Target')
target_lb.grid(row=3, column=0)

last5_lb = Label(root,text='Enter Last 5 Over Runs')
last5_lb.grid(row=4, column=0)

batteam_lb = Label(root,text='Enter Batting Team')
batteam_lb.grid(row=5, column=0)

bowlteam_lb = Label(root,text='Enter Bowling Team')
bowlteam_lb.grid(row=6, column=0)

venue_lb = Label(root,text='Enter Venue')
venue_lb.grid(row=7, column=0)

predict_lb = Label(root,text='Prediction : ')
predict_lb.grid(row=9,column=0)

prediction_lb = Label(root)
prediction_lb.grid(row=9,column=1)

run_in = Entry(root)
run_in.grid(row=0, column=1)

wicket_in = Entry(root)
wicket_in.grid(row=1, column=1)

ball_in = Entry(root)
ball_in.grid(row=2, column=1)

target_in = Entry(root)
target_in.grid(row=3, column=1)

last5_in = Entry(root)
last5_in.grid(row=4, column=1)

batteam_om = ttk.OptionMenu(root,batteam_select,'Select Team',*team_option)
batteam_om.grid(row=5, column=1)

ballteam_om = ttk.OptionMenu(root,ballteam_select,'Select Team',*team_option)
ballteam_om.grid(row=6, column=1)

venue_cb = ttk.Combobox(root,textvariable=venue_select,values=venue_option)
venue_cb.bind("<KeyRelease>", update_opt)
venue_cb.grid(row=7, column=1)

predict_btn=Button(root, text='Predict', command=predict_comm)
predict_btn.grid(row=8, column=1)


root.mainloop()
```
