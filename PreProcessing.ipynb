```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df = df.loc[df.Innings == 2]
df
```

```python
df.reset_index()
```

```python
df['Balls_left'] = 126 - (6*df['Over'] + df['Ball'])
df['Balls_bowled'] = 120 - df['Balls_left'] 
df['Runs_left'] = df['Target Score'] - df['Innings Runs']
```

```python
df.head()
```

```python
# remove the rows with infinte RRR
df = df.loc[df['Balls_left']!=0]
df = df.reset_index()
```

```python
df.head()
```

```python
dataset = df.drop(['Match ID', 'Innings', 'Over', 'Ball'], axis = 1)
# randomly distributing data
dataset = dataset.sample(n=len(dataset))
dataset = dataset.reset_index(drop = True)
dataset = dataset.drop('index' ,axis =1)
dataset
```

```python
# One Hot Encoding
final = pd.get_dummies(dataset, columns = ['Venue', 'Bat First', 'Bat Second'])
final = final*1
final
```

```python
final.to_csv("Data.csv")
```
