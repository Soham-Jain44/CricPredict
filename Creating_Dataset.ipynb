```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('ball_by_ball_it20.csv')
```

```python
data
```

```python
newdata = data.drop(columns=['Date','Batter','Non Striker','Bowler','Extra Runs','Runs From Ball','Ball Rebowled','Extra Type','Wicket','Method','Player Out','Winner','Total Batter Runs','Total Non Striker Runs','Batter Balls Faced','Non Striker Balls Faced','Player Out Runs','Player Out Balls Faced','Bowler Runs Conceded','Valid Ball'])
```

```python
eligible_venues = newdata.groupby('Venue')['Venue'].value_counts()[newdata.groupby('Venue')['Venue'].value_counts()>600].index.tolist()
```

```python
newdata = newdata[newdata['Venue'].isin(eligible_venues)]
```

```python
newdata['Bat First'].unique()
```

```python
eligible_teams = ['West Indies',
       'Sri Lanka',
        'England',
       'Bangladesh',
       'India',
        'Afghanistan',
        'New Zealand',
        'Pakistan',
       'South Africa',
        'Australia',]
```

```python
newdata = newdata[newdata['Bat First'].isin(eligible_teams)]
```

```python
newdata = newdata[newdata['Bat Second'].isin(eligible_teams)]
```

```python
matches = newdata.groupby(['Match ID','Innings'])
ls = []
for id in newdata['Match ID'].unique():
    for inn in range(1,3):
        ls.extend(matches.get_group((id,inn))['Batter Runs'].rolling(window=30).sum().tolist())
```

```python
newdata['Last Five'] = ls
```

```python
newdata['CRR'] = (newdata['Innings Runs']*6)/(newdata['Over']*6+newdata['Ball'])
```

```python
newdata=newdata.dropna(subset=['Last Five'])
```

```python
newdata['RRR'] = (newdata['Runs to Get']*6)/newdata['Balls Remaining']
```

```python
newdata.info()
```

```python
newdata['Wickets Left'] = 10-newdata['Innings Wickets']
```

```python
newdata = newdata.drop(columns=['Unnamed: 0','Batter Runs','Innings Wickets','Runs to Get','Balls Remaining'])
```
